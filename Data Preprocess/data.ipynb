```python
Schools = []
for i in range(len(uni_faculties)):
    for j in range(len(uni_departments[i])):
        Schools.extend([hk_universities[i]] * len(uni_departments[i][j]))
Faculties = []
for i in range(len(uni_faculties)):
    for j in range(len(uni_departments[i])):
        Faculties.extend([uni_faculties[i][j]] * len(uni_departments[i][j]))
Departments = []
for i in range(len(uni_faculties)):
    for j in range(len(uni_departments[i])):
        for k in range(len(uni_departments[i][j])):
            Departments.append(uni_departments[i][j][k])
Links = []
for i in range(len(uni_faculties)):
    for j in range(len(uni_departments[i])):
        for k in range(len(uni_departments[i][j])):
            if (i == 1 or i == 2):
                Links.append(uni_links[i])
            else:
                if (type(hku_links[j]) == str):
                    Links.append(hku_links[j])
                else:
                    Links.append(hku_links[j][k])
```

```python
# All in one
df["Schools"] = Schools
df["Faculties"] = Faculties
df["Department"] = Departments
df["Links"] = Links

df
```
